```python
import pandas as pd

safety_mean = df['safety'].mean()
safety_median = df['safety'].median()
safety_mode = df['safety'].mode()[0]
print(f"Safety Column:")
print(f"  Mean: {safety_mean}")
print(f"  Median: {safety_median}")
print(f"  Mode: {safety_mode}")

buying_mean = df['buying'].mean()
buying_median = df['buying'].median()
buying_mode = df['buying'].mode()[0]

print(f"\nBuying Column:")
print(f"  Mean: {buying_mean}")
print(f"  Median: {buying_median}")
print(f"  Mode: {buying_mode}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd


combined_counts = pd.concat([df['safety'].value_counts().sort_index(),
                             df['buying'].value_counts().sort_index()], axis=1)
combined_counts.columns = ['Safety', 'Buying']

ax = combined_counts.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightcoral'])

plt.title('Distribution of Safety and Buying Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Column')
plt.tight_layout()
plt.show()
```
